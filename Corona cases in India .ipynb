```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.mohfw.gov.in/"
```

```python
response = requests.get(url)
print(response)
```

```python
soup = BeautifulSoup(response.content,"html.parser")
```

```python
divs = soup.find("div" ,class_="container")
for i in divs.find_all("div",class_="update-box"):
    print(i.text)
```

```python
table = soup.find_all("table")
len(table)
```

```python
table = table[0]
```

```python
Sr_no=[]
State_or_UT=[]
Total_cases=[]
Discharged=[]
Deaths=[]

rows = table.find_all("tr")[1:33]
for row in rows:

        col = row.find_all("td")
        Sr_no.append(col[0].text.strip())
        State_or_UT.append(col[1].text.strip())
        Total_cases.append(col[2].text.strip())
        Discharged.append(col[3].text.strip())
        Deaths.append(col[4].text.strip())
        
rows = table.find_all("tr")[33:34]
print(rows)
for row in rows:
    col=row.find_all("strong")
    State_or_UT.append(col[0].text.strip())
    Total_cases.append(col[1].text.strip())
    Discharged.append(col[2].text.strip())
    Deaths.append(col[3].text.strip())

print(State_or_UT)
print(Deaths)
print(Discharged)
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame(list(zip(Sr_no,State_or_UT,Total_cases,Discharged,Deaths)),columns=["Sr.no","State/UT","Total_cases","Discharged","Deaths"])
data.set_index("Sr.no")
```

```python
State_or_UT1=[]
for i in data["State/UT"]:
    s1=str(i)
    s2=s1.replace("#","")
    
    State_or_UT1.append(s2)
State_or_UT =State_or_UT1

Deaths1=[]
for i in data["Deaths"]:
    s1=str(i)
    s2=s1.replace("#","")
    s3=int(s2)
    Deaths1.append(s3)
Deaths = Deaths1
```

```python
data = pd.DataFrame(list(zip(Sr_no,State_or_UT,Total_cases,Discharged,Deaths)),columns=["Sr.no","State/UT","Total_cases","Discharged","Deaths"])
data.set_index("Sr.no")
```
